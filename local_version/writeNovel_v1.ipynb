```python
# from google.colab import userdata
import getpass
import os
import google.generativeai as genai
```

```python
# 실행하기 전에 열쇠창에 API키를 등록하세요.

# 코랩버전이 아니므로 주석처리 로컬 환경변수에서 불러오는 것으로 변경
'''
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
'''

# Gemini 설정
genai.configure(api_key=os.environ.get("GOOGLE_API_KEY"))
model = genai.GenerativeModel(model_name = "gemini-pro")
```

```python
# 종종 실행이 안됩니다. 그럴경우 런타임 연결해제 및 재시작 후 다시 시도하세요.
import time
import re
from tqdm.notebook import tqdm
from google.generativeai import types as genai

def generate_selected_sentences(text):
    sentences = text.split('.')
    selected_sentences = ""
    for index in [-50, -40, -35, -30, -25, -20, -16, -12, -8, -4, 1]:
        try:
            selected_sentences = '.'.join(sentences[index::-1]).strip()
            break
        except IndexError:
            continue
    if not selected_sentences:
        selected_sentences = '.'.join(sentences[1:])
    return selected_sentences

prompt = "소프트웨어 개발자를 준비하는 청년들이 있었습니다. 이들은 새싹에서 주관하는 구글클라우드인공지능개발자 과정을 수강중입니다. "
prompt2 = """{당신은 한국의 매우 뛰어난 실력을 갖춘 저명한 소설 작가입니다. 이어지는 소설을 길게 작성해주세요.
스토리가 아주 조금씩 전개되도록 작성해주세요. 아래의 모든 조건들을 반드시 반드시 반드시!! 지켜주세요.
조건1 : 영어 사용금지.
조건2 : 중국어 사용금지.
조건3 : 숫자 사용금지.
조건4 : 스토리가 이어질 수 있는 문장으로 마무리 하시오.
조건5 : 마지막에 구두점(.)으로 문장을 마무리 하시오.
조건6 : 'HARM_CATEGORY'에 속하는 문장을 절대로 생성하지 마시오.
조건7 : 도덕적이지 않은내용, 폭력적인 내용을 절대로 생성하지 마시오.
조건7 : 스토리의 결말을 지어서는 안됨.
조건8 : 줄바꿈을 적절히 활용하시오.
조건9 : 주인공은 완규입니다.
조건10 : 동료는 규미, 지호, 창석, 시내입니다.
조건11 : 대명사 사용을 자제하고, 주인공의 이름을 지속적으로 사용하시오.
조건12 : 주인공과 친구들은 절대 사망해서는 안됩니다.
조건13 : 인물들간에 대화를 자주 삽입 하시오.}"""
prompt3 = "조건 : 진행되고 있던 소설을 마무리 하세요."
n = 100
myNovel = ""

for i in tqdm(range(n), desc="Generating Text"):
    current_prompt = prompt3 if i == n - 1 else prompt2 + prompt
    try:
        generation_config = genai.GenerationConfig(temperature=1.0)
        response = model.generate_content(current_prompt, generation_config=generation_config)
        if not response.candidates:
            print(f"프롬프트 {i}에 대한 유효한 후보가 생성되지 않았습니다. response.prompt_feedback를 확인하세요.")
            print(f"프롬프트 피드백: {response.prompt_feedback}")
            continue

        generated_text = response.text
        if not generated_text:
            print(f"응답에 텍스트가 포함되지 않았습니다. 다음 반복으로 건너뜁니다.")
            continue

        if re.search("[a-zA-Z]", generated_text):
            print(f"생성된 텍스트에 영어 문자가 포함되어 있습니다. 다음 반복으로 건너뜁니다.")
            continue

        selected_sentences = generate_selected_sentences(generated_text)
        if not selected_sentences:
            print(f"문장 선택에 실패했습니다. 다음 반복으로 건너뜁니다.")
            continue

        prompt = selected_sentences + '\n'
        myNovel += generated_text
        # time.sleep(5)
        # 사용제한이 계속 걸린다면 위의 time.sleep 코드를 사용하세요.
    except ValueError as e:
        continue

# 오류가 수차례 연속적으로 반복될경우 실행을 중단하세요.
# 종종 실행이 안됩니다. 그럴경우 런타임 연결해제 및 삭제 후 다시 시도하세요.
```
